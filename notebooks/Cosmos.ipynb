```python
import random

SEED = 1122334455
STAR_COLOR = ["RED", "ORANGE", "YELLOW", "BLUE", "WHITE"]
MAIN_MASS = [6, 7, 7, 8, 8, 9]
MAIN_ORBI = [
    [0.2, 0.3, 0.4, 0.5, 0.7, 0.9, 1.1, 1.4, 1.7, 2],
    [2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0],
    [7.0, 8.0, 9.0, 10.0, 11.0, 12.0],
]

class Planet(object):
    def __init__(self, mass, orbit):
        self.mass = mass
        self.orbit = orbit
        self.terrain = 0
        
    def __repr__(self):
        return "M: %d, O: %f" % (self.mass, self.orbit)
    
    
class StarSystem(object):
    def __init__(self, seed):
        self.rng = random.Random()
        self.rng.seed(seed)
        self.color = self.rng.randint(0, len(STAR_COLOR))
        self.planets = [0] * self.rng.randint(2, 14)
        mo_pos = self.rng.randint(1,6)
        if mo_pos < 3:
            mo_pos = 0
        elif mo_pos < 6:
            mo_pos = 1
        else:
            mo_pos = 2
        if mo_pos == 0:
            self.main = self.rng.randint(1, 4) - 2
        else:
            self.main = self.rng.randint(1, 4) + 1
        if self.main >= (len(self.planets) - 1):
            self.main = len(self.planets) - 1
        if self.main < 0:
            self.main = 0
        self.planets[self.main] = Planet(self.rng.choice(MAIN_MASS), self.rng.choice(MAIN_ORBI[mo_pos]))
        
    def print(self, indent=""):
        print("%sStar color: %s" % (indent, STAR_COLOR[self.color]))
        print("%sPlanets: %d, %d inner" % (indent, len(self.planets), self.main))
        print("%sMain: %s" % (indent, repr(self.planets[self.main])))

```

```python
sys = StarSystem(11223344)
sys.print()
```

```python
seed = 0
while True:
    sys = StarSystem(seed)
    p = sys.planets[sys.main]
    if (sys.color == 1 or sys.color == 2) and sys.main > 1 and p.mass >7 and p.orbit > 1.0:
        print("Seed: ", seed)
        sys.print("  ")
        break
    seed += 1
```
