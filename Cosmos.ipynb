```python
import random as rand

SEED = 1122334455
STAR_COLOR = ["RED", "ORANGE", "YELLOW", "BLUE", "WHITE"]
MAIN_MASS = [6, 7, 7, 8, 8, 9]
MAIN_ORBI = [
    [0.2, 0.3, 0.4, 0.5, 0.7, 0.9, 1.1, 1.4, 1.7, 2],
    [2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0],
    [7.0, 8.0, 9.0, 10.0, 11.0, 12.0],
]

rand.seed(SEED)
print("Star color: %s" % rand.choice(STAR_COLOR))
plnts = rand.randint(2,14)
print("Planets: %d" % plnts)
print("Mass of main: %d" % rand.choice(MAIN_MASS))
mo_pos = rand.randint(1,6)
if mo_pos < 3:
    mo_pos = 0
elif mo_pos < 6:
    mo_pos = 1
else:
    mo_pos = 2
print("Orbit of main: %f" % rand.choice(MAIN_ORBI[mo_pos]))
if mo_pos == 0:
    ipn = rand.randint(1, 4) - 2
else:
    ipn = rand.randint(1, 4) + 1
if ipn >= (plnts - 1):
    ipn = plnts - 1
if ipn > 0:
    print("Inner planets: %d" % ipn)
```
